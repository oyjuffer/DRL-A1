```python
from sklearn.model_selection import ParameterGrid
import pandas as pd

parameters = {'batch_size': [32,64], 'epsilon': [1], 'epsilon_end': [0.1],
              'epsilon_decay': [0.001], 'gamma': [0.95, 0.99], 
              'n_episodes': [2000, 3000, 4000],
              'n_evaluate': [10], 'n_update': [50, 100, 500], 
              'learning_rate': [0.00025, 0.0005]
              }

cases = list(ParameterGrid(parameters))

for case, c_iter in zip(cases, range(len(cases))):
    output_name = 'case_{}'.format(c_iter)
    hyperparams_file = output_name + ".xlsx"
    pd.DataFrame(case, index=[0]).to_excel("params/{}".format(hyperparams_file))
```
